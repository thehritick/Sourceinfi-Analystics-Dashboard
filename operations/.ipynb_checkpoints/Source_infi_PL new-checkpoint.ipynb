```python
#source-infi
db_connection_str1 = 'mysql+pymysql://admin:Qazwsx123@sourceinfi-read-replica.clwjnp7sjnpx.ap-south-1.rds.amazonaws.com/voehoo'
db_connection = create_engine(db_connection_str1)

sql1='select s.sp_shipping_id as Nimbus_Shipping_ID, sot.variant_sku as variant_sku, sot.b2b_price as b2b_price, sot.selling_price as selling_price, sot.rto_margin as rto_margin from shippings as s LEFT JOIN shopify_order_items as sot on s.order_item_id = sot.id LEFT JOIN shopify_orders as so on so.id = sot.order_id  left join users as u on u.id=s.user_id where concat(u.fname ," ", u.lname) <> "Vendor Test" '
scinfi_data=pd.read_sql(sql1, con=db_connection)
scinfi_data=scinfi_data[scinfi_data.Nimbus_Shipping_ID!=0]
scinfi_data.reset_index(inplace=True,drop=True)

sku_data=nimbus_data.merge(scinfi_data,on='Nimbus_Shipping_ID', how='left',suffixes=('_SI', '_Nimbus'),indicator=True)
sku_data[['b2b_price','selling_price','rto_margin']]= sku_data[['b2b_price','selling_price','rto_margin']].astype(float)

sku_data['margin']=sku_data['selling_price']-sku_data['b2b_price']
scinfi_data
```

```python
###########

x=sku_data.pivot_table(
                        values=['margin','rto_margin'],
                        index='variant_sku',
                        aggfunc='mean',
                        fill_value=0,
                        margins=True,
                        margins_name='Grand_Total'
                        ).reset_index()
x.rename(columns={'margin':'avg_margin','rto_margin':'avg_rto_margin'},inplace=True)

cost=sku_data.pivot_table(
                        values='final_charges',
                        index='variant_sku',
                        aggfunc='sum',
                        fill_value=0,
                        margins=True,
                        margins_name='Grand_Total'
                        ).reset_index()

c_margin_rev=sku_data.pivot_table(
                            values='margin',
                            index='variant_sku',
                            columns='Nimbus_Tracking_Status',
                            aggfunc='sum',
                            fill_value=0,
                            margins=True,
                            margins_name='Grand_Total'
                            ).reset_index()
c_margin_rev=c_margin_rev[['variant_sku','delivered']]

c_rtomargin_rev=sku_data.pivot_table(
                            values='rto_margin',
                            index='variant_sku',
                            columns='Nimbus_Tracking_Status',
                            aggfunc='sum',
                            fill_value=0,
                            margins=True,
                            margins_name='Grand_Total'
                            ).reset_index()
c_rtomargin_rev=c_rtomargin_rev[['variant_sku','in transit','out for delivery','exception','rto']]

c_rev=c_margin_rev.merge(c_rtomargin_rev,on='variant_sku')
c_rev['total_rev']=c_rev['delivered']+c_rev['in transit']+c_rev['out for delivery']+c_rev['exception']+c_rev['rto'] 

metrix=sku_data.pivot_table(
                            values='awb_number',
                            index='variant_sku',
                            columns='Nimbus_Tracking_Status',
                            aggfunc='count',
                            fill_value=0,
                            margins=True,
                            margins_name='Grand_Total'
                            ).reset_index()
metrix.rename(columns={'Grand_Total':'Total_Shipments'},inplace=True)

metrix['del_perc']= round((metrix['delivered']/ (metrix['delivered']+metrix['in transit']+metrix['out for delivery']+metrix['exception']+metrix['rto'])),2)

final=metrix.merge(c_rev.merge(cost.merge(x,on='variant_sku'),on='variant_sku'),on='variant_sku',suffixes=['_count','_margin'])
final['Live_PnL']=final['total_rev']-final['final_charges']
final['rto_perc']=1-final['del_perc']
final['Projected_rev']=(final['Total_Shipments']*final['del_perc']*final['avg_margin'])+(final['Total_Shipments']*final['rto_perc']*final['avg_rto_margin'])
final['Projected_PnL']=final['Projected_rev']-final['final_charges']

final
```
