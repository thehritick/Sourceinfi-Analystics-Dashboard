```python
#source-infi
db_connection_str1 = 'mysql+pymysql://admin:Qazwsx123@sourceinfi-read-replica.clwjnp7sjnpx.ap-south-1.rds.amazonaws.com/voehoo'
db_connection1 = create_engine(db_connection_str1)
sql1='select s.sp_shipping_id as Nimbus_Shipping_ID, sot.variant_sku as variant_sku, sot.b2b_price as b2b_price, sot.selling_price as selling_price, sot.rto_margin as rto_margin from shippings as s LEFT JOIN shopify_order_items as sot on s.order_item_id = sot.id LEFT JOIN shopify_orders as so on so.id = sot.order_id  left join users as u on u.id=s.user_id where concat(u.fname ," ", u.lname) <> "Vendor Test" '
scinfi_data=pd.read_sql(sql1, con=db_connection1)
scinfi_data

scinfi_data=scinfi_data[scinfi_data.Nimbus_Shipping_ID!=0]
scinfi_data.reset_index(inplace=True,drop=True)
scinfi_data
```

```python
sku_data=nimbus_data.merge(scinfi_data,on='Nimbus_Shipping_ID', how='left',suffixes=('_SI', '_Nimbus'),indicator=True)
# sku_data= sku_data.fillna({'b2b_price': 0, 'selling_price': 0, 'rto_margin':0})
sku_data[['b2b_price','selling_price','rto_margin']]= sku_data[['b2b_price','selling_price','rto_margin']].astype('float64')
sku_data
```

```python
sku_data['margin']=sku_data['selling_price']-sku_data['b2b_price']
sku_data
```

```python
sku_data.final_charges.sum()
```

```python
del_shipments=sku_data[sku_data.Nimbus_Tracking_Status=='delivered']
del_shipments

del_perf= del_shipments.groupby('variant_sku').agg({'margin': 'sum'}).reset_index()
del_perf
```

```python
rto_shipments=sku_data[sku_data.Nimbus_Tracking_Status=='rto']
rto_shipments

rto_perf= rto_shipments.groupby('variant_sku').agg({'rto_margin': 'sum'}).reset_index()
rto_perf
```

```python
final_charges= sku_data.groupby('variant_sku').agg({'final_charges': 'sum'}).reset_index()
final_charges
```

```python
final_charges.final_charges.sum()
```

```python
metrix=sku_data.pivot_table(
                            values='awb_number',
                            index='variant_sku',
                            columns='Nimbus_Tracking_Status',
                            aggfunc='count',
                            fill_value=0,
                            margins=True,
                            margins_name='Total_Shipments'
                            ).reset_index()

metrix= metrix[metrix.variant_sku != 'Total_Shipments']

metrix=metrix.merge(del_perf, on='variant_sku', how='outer')
metrix=metrix.merge(rto_perf, on='variant_sku', how='outer')
metrix=metrix.merge(final_charges, on='variant_sku', how='outer')

metrix=metrix.fillna(0)

# metrix['del_perc']= round((metrix['delivered']/ (metrix['delivered']+metrix['exception']+metrix['rto']))*100,2)
# metrix.del_perc= metrix.del_perc.astype('str') + '%'

# metrix['live_PL']= (metrix.margin + metrix.rto_margin) - metrix['final_charges']
# metrix['PL_per_ship']= round(metrix.live_PL / metrix.Total_Shipments)

metrix
```
